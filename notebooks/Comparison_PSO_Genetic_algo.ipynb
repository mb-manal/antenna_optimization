```python
import json
from pathlib import Path
import pandas as pd
import matplotlib.pyplot as plt
```

```python

OUT = Path("../outputs")

with open(OUT / "metrics_ga.json") as f:
    ga = json.load(f)
with open(OUT / "metrics_pso.json") as f:
    pso = json.load(f)

df = pd.DataFrame([ga, pso]).set_index("algo")[["SLL_dB","FWHM_x","FWHM_y","FWHM_sum","Eccentricity","Score"]]
display(df)

# Bar chart comparatif
plt.figure(figsize=(8,4))
df[["SLL_dB","FWHM_sum","Eccentricity"]].plot(kind="bar")
plt.title("GA vs PSO â€” SLL / FWHM_sum / Eccentricity")
plt.ylabel("valeur (SLL dB, FWHM px, ecc)")
plt.grid(True, axis='y')
plt.tight_layout(); plt.show()
```
